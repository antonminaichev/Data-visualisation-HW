Определим средний, максимальный и минимальный возраст жертв - атрибут victim_age. Для этого сначала уберем все записи со значением возраста Unknown. По оставшимся записям определим необходимые значения.

```python
df_age = df.query("victim_age != 'Unknown'")
avg_age = df_age['victim_age'].astype(float).mean()
max_age = df_age['victim_age'].astype(float).max()
min_age = df_age['victim_age'].astype(float).min()
data_age = {'Средний возраст':  [avg_age],
        'Максимальный возраст': [max_age],
         'Минимальный возраст': [min_age]
        }

age_table = pd.DataFrame(data_age)

print(age_table)
```

Из данных видно, что средний возраст жертвы - 31.8 лет, самой старой жертве - 102 года, самой молодой - было меньше года.

Посчитаем количество мужчин и женщин среди жертв.

```python
man_count = df['victim_sex'].value_counts()['Male']
female_count = df['victim_sex'].value_counts()['Female']
unknw_count = df['victim_sex'].value_counts()['Unknown']
man_percentage = man_count / df['victim_sex'].count() * 100
female_percentage = female_count / df['victim_sex'].count() * 100
unknw_percentage = unknw_count / df['victim_sex'].count() * 100
data_sex = {'Количество': [man_count, female_count, unknw_count],
        '%': [man_percentage, female_percentage, unknw_percentage]
        }

sex_table = pd.DataFrame(data_sex, index=['Man', 'Woman', 'Unknown'])

print(sex_table)
```

Из данных видно, что большинство жертв - мужчины, а также что для ~8% жертв - пол скрыт или неизвестен. 

Посчитаем количество женщин, которым больше 18 лет.

```python
df_age['victim_age'] = df_age['victim_age'].astype(float)
df_female = df_age.query("victim_sex == 'Female' & victim_age > 18")
female_18_count = df_female['uid'].count()
print("Количество женщин жертв старше 18 лет -", female_18_count, '\nВ подсчете не учитываются те записи, где пол жертвы скрыт или неизвестен.')
```

Подсчитаем количество мужчин-жертв, которые были темнокожими.

```python
df_male = df.query("victim_sex == 'Male'")
black_man_count = df_male['victim_race'].value_counts()['Black']
print("Количество темнокожих мужчин-жертв -", black_man_count, '\nВ подсчете не учитываются те записи, где раса жертвы скрыта или неизвестна.')
black_man_percent_total = black_man_count / df['victim_sex'].count() * 100
black_man_percent_man = black_man_count / man_count * 100
print("Процент темнокожих мужчин-жертв среди всех жертв -", black_man_percent_total, '%', '\nПроцент темнокожих мужчин-жертв среди всех жертв-мужчин - ', black_man_percent_man, '%')
```

Таким образом, больше всего жертв - среди темнокожих мужчин Америки. Почему имеем такое распределение - тема для размышления...
